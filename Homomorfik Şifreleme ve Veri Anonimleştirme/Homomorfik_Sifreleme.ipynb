```python
from Crypto.Util import number

# Veri seti oluşturma (örneğin 42 elemanlı rastgele bir dizi)
data_set = [number.getRandomRange(1, 1000) for _ in range(42)]

# ElGamal şifreleme işlemi
def encrypt(data, p, g, public_key):
    encrypted_data = []
    for d in data:
        k = number.getRandomRange(1, p - 2)  # Rastgele bir k değeri seç
        c1 = pow(g, k, p)
        c2 = (d * pow(public_key, k, p)) % p
        encrypted_data.append((c1, c2))
    return encrypted_data

# ElGamal deşifreleme işlemi
def decrypt(encrypted_data, private_key, p):
    decrypted_data = []
    for c1, c2 in encrypted_data:
        s = pow(c1, private_key, p)
        decrypted = (c2 * pow(s, -1, p)) % p
        decrypted_data.append(decrypted)
    return decrypted_data

# Anahtar oluşturma
def generate_keys():
    p = number.getPrime(128)  # Rastgele bir asal sayı seç
    g = number.getRandomRange(2, p - 1)  # Rastgele bir g değeri seç
    private_key = number.getRandomRange(2, p - 1)  # Rastgele bir özel anahtar seç
    public_key = pow(g, private_key, p)  # Açık anahtar hesapla
    return p, g, public_key, private_key

# Anahtarları oluştur
p, g, public_key, private_key = generate_keys()

# Veriyi şifrele
encrypted = encrypt(data_set, p, g, public_key)
print("Şifrelenmiş Veri:")
print(encrypted)

# Veriyi deşifre et
decrypted = decrypt(encrypted, private_key, p)
print("\nDeşifre Edilmiş Veri:")
print(decrypted)
```

```python
from pycryptodome import Random
from pycryptodome.math.numbers import Integer


# Veri seti
veri = {
    'TC': [12345612345, 65432145678, 74185298765, 98768536985, 75395102587, 15948600124, 85296312025, 25636936984,
            14747112302, 25852834821, 48585217935, 36636332069, 56556551452, 78778777889, 98566915415,
            52520100025, 10203036063, 10050391515, 40506075398, 78009921843]
}


# Bir homomorfik şifreleme fonksiyonu oluşturalım
def encrypt(value, key):
    return value * key  # Bu örnekte, çarpma işlemiyle şifreleme yapıyoruz


# Deşifreleme fonksiyonu
def decrypt(value, key):
    return value // key  # Şifreli değeri anahtarla bölerek orijinal değeri elde ediyoruz


# Anahtar oluşturalım (rastgele bir sayı)
encryption_key = Random.get_random_integer(128)


# Veri setindeki her TC kimlik numarasını şifreleyelim
encrypted_data = {}
for tc in veri['TC']:
    encrypted = encrypt(tc, encryption_key)
    encrypted_data[tc] = encrypted
    

# Şifrelenmiş veriyi gösterelim
print("Şifrelenmiş Veri:")
for tc, encrypted in encrypted_data.items():
    print(f"TC: {tc}, Şifreli: {encrypted}")
    

# Şifreli veriyi deşifre edelim ve orijinal veriyi elde edelim
decrypted_data = {}
for tc, encrypted in encrypted_data.items():
    decrypted = decrypt(encrypted, encryption_key)
    decrypted_data[encrypted] = decrypted
    

# Deşifre edilmiş veriyi gösterelim
print("\nDeşifre Edilmiş Veri:")
for encrypted, decrypted in decrypted_data.items():
    print(f"Şifreli: {encrypted}, Deşifreli TC: {decrypted}")
```

```python
# Homomorfik çarpma işlemi için örnek
def homomorphic_encrypt(value):
    # Şifreleme işlemi: Değerin 2 ile çarpılması
    encrypted_value = value * 2  # Çarpma işlemi şifrelemeyi temsil ediyor
    return encrypted_value

def homomorphic_decrypt(value):
    # Şifreli değerin orijinal değere dönüştürülmesi: Değerin 2'ye bölünmesi
    decrypted_value = value // 2  # Bölme işlemi deşifrelemeyi temsil ediyor
    return decrypted_value

# Veri seti
veri = {
    'TC': [12345612345, 65432145678, 74185298765, 98768536985, 75395102587, 15948600124, 85296312025, 25636936984,
            14747112302, 25852834821, 48585217935, 36636332069, 56556551452, 78778777889, 98566915415,
            52520100025, 10203036063, 10050391515, 40506075398, 78009921843]
}

# Homomorfik şifreleme işlemi
encrypted_data = {}
for tc in veri['TC']:
    encrypted_value = homomorphic_encrypt(tc)
    encrypted_data[tc] = encrypted_value

# Şifrelenmiş veriyi gösterelim
print("Şifrelenmiş Veri:")
for tc, encrypted in encrypted_data.items():
    print(f"TC: {tc}, Şifreli: {encrypted}")

# Deşifreleme işlemi
decrypted_data = {}
for tc, encrypted in encrypted_data.items():
    decrypted_value = homomorphic_decrypt(encrypted)
    decrypted_data[encrypted] = decrypted_value

# Deşifre edilmiş veriyi gösterelim
print("\nDeşifre Edilmiş Veri:")
for encrypted, decrypted in decrypted_data.items():
    print(f"Şifreli: {encrypted}, Deşifreli TC: {decrypted}")
```
